```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/DSData/master/happiness_score_dataset.csv')
```

```python
data
```

```python
data.describe()
```

```python
data.isna().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc = LabelEncoder()
```

```python
df = lab_enc.fit_transform(data['Country'])
```

```python
pd.Series(df)
```

```python
data['Country'] = df
```

```python
df1 = lab_enc.fit_transform(data['Region'])
pd.Series(df1)
```

```python
data['Region'] = df1
```

```python
data
```

```python
data.describe()
```

```python
df = data.corr()
plt.figure(figsize=(12,14))
sn.heatmap(df,annot=True)
plt.show()
```

```python
x = data.drop(columns=['Happiness Score'])
y = data['Happiness Score']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state = 77)
```

```python
 from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn import metrics
```

```python
abr = AdaBoostRegressor()
```

```python
abr.fit(x_test,y_test)
```

```python
y_pred = abr.predict(x_test)
```

```python
metrics.r2_score(y_test,y_pred)
```
